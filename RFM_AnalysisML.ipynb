# Recency

```python
R_data = data.loc[ : ,['Customer_ID', 'Customer_Name','Order_Date']]
```

```python
R_data.head(10)
```

```python
R_data.info()
```

```python
R_data['Order_Date'] = pd.to_datetime(R_data['Order_Date'], format='%d-%m-%Y')
```

```python
R_data.info()
```

```python
R_data.head()
```

```python
R_data = R_data.sort_values(by = ['Customer_ID','Order_Date'], ascending = [True,False])
```

```python
R_data.head(20)
```

```python
from datetime import date
```

```python
current_date = date.today()
print(current_date)
```

```python
R_data['present_date'] = current_date
```

```python
R_data.head()
```

```python
R_data.info()
```

```python
R_data['present_date'] = pd.to_datetime(R_data['present_date'])
```

```python
R_data['no_of_days'] = R_data['present_date'] - R_data['Order_Date']
```

```python
R_data.head()
```

```python
R_data.info()
```

```python
#How to convert a timedel
R_data['no_of_days'] = R_data['no_of_days'].replace("days","").astype('int64')
```

```python
R_data.head()
```

```python
# Keep only the most recent transaction for each customer
recency_data = R_data.groupby('Customer_ID').first().reset_index()
```

## frequency

```python
# Calculate frequency by counting the number of orders for each customer
frequency_data = R_data.groupby('Customer_ID').agg({'Order_Date': 'count'}).reset_index()
frequency_data.columns = ['Customer_ID', 'frequency']
```

```python
# Merge with recency data
RFM_data = pd.merge(recency_data, frequency_data, on='Customer_ID')
```

```python
RFM_data.head(20)
```

## monetary

```python
# Assuming you have 'Sales' column in your data representing purchase amount
monetary_data = data.groupby('Customer_ID').agg({'Sales': 'sum'}).reset_index()
monetary_data.columns = ['Customer_ID', 'monetary']
```

```python
RFM_data.columns
```

```python
# Merge with the existing RFM data
RFM_data = pd.merge(RFM_data, monetary_data, on='Customer_ID')

# Display the RFM data for the first 10 customers
RFM_data[['Customer_ID', 'Customer_Name', 'no_of_days', 'frequency', 'monetary']].head(10)
```

```python
RFM_data.head(20)
```

```python
# Create quartile-based scoring for Recency, Frequency, and Monetary
RFM_data['R_Score'] = pd.qcut(RFM_data['no_of_days'], 4, labels=[4, 3, 2, 1])
RFM_data['F_Score'] = pd.qcut(RFM_data['frequency'].rank(method='first'), 4, labels=[1, 2, 3, 4])
RFM_data['M_Score'] = pd.qcut(RFM_data['monetary'].rank(method='first'), 4, labels=[1, 2, 3, 4])
```

```python
RFM_data.head(20)
```

```python
# Combine scores to get the RFM score
RFM_data['RFM_Score'] = RFM_data['R_Score'].astype(str) + RFM_data['F_Score'].astype(str) + RFM_data['M_Score'].astype(str)
```

```python
# Show the RFM data with RFM scores for first 10 customers
RFM_data[['Customer_ID', 'Customer_Name', 'RFM_Score']].head(10)
```

```python
# Define customer segments based on RFM Score
def segment_customer(df):
    if df['RFM_Score'] == '444':
        return 'Best Customers'
    elif df['RFM_Score'].startswith('4'):
        return 'Loyal Customers'
    elif df['RFM_Score'].startswith('1'):
        return 'At Risk'
    else:
        return 'Others'

RFM_data['Customer_Segment'] = RFM_data.apply(segment_customer, axis=1)

# Show segmented customer data
RFM_data[['Customer_ID', 'Customer_Name', 'RFM_Score', 'Customer_Segment']].head(10)
```
